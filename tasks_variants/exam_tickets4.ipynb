
## Завдання на практичну частину екзамену з дисципліни МІАЗ.

### Білет 4

Створити прототип інформаційно-аналітичної системи для відображення інформації щодо втрат противника у вантажівках за останні 15 днів на основі наданого шаблону.

1. Створити базу даних exam_60, створити таблицю та заповнити її даними за допомогою sql-файлів з шаблону.
2. Backend на основі Python FastAPI (налаштувати параметри підключення до бази даних, та прописати оператор SELECT для вибірки необхідних полів: 
   date, location, [поле відповідно до завдання] as data_field)
3. Frontend на основі:
   - HTML(додати місця розташування відповідних візуальних елементів)
   - CSS (налаштувати кольорову схему)
   - JavaScript (встановити базовий колір графіку та задати координати маркерів)

4. Кольорова схема має відповідати основному кольору Степ відповідно до [брендбуку ЗСУ](https://sprotyvg7.com.ua/wp-content/uploads/2024/06/1_%D0%9A%D0%95%D0%A0-%D0%97-%D0%92%D0%98%D0%9A-%D0%93%D0%A0-%D0%A1%D0%A2%D0%98%D0%9B%D0%AE-%D0%A3-%D0%92%D0%86%D0%97-%D0%9A%D0%9E%D0%9C-%D0%97%D0%A1%D0%A3.pdf).
5. Координати маркерів на інтерактивній карті:
     A: [50.9919, 34.7878]
     B: [48.4654, 35.0319]
     C: [46.6523, 32.6286] 
6. Запустити backend (uvicorn main:app --reload) та frontend (live server) частини інформаційно-аналітичної системи.
7. Надати роботу на перевірку.
